```python
volcano_summary_list=[]
```

```python
volcano_types = ['Caldera', 'Complex volcano', 'Volcanic cone', 'Explosion crater', 'Fissure vent', 'Lava cone', 'Lava dome', 'Maar', 'Pyroclastic shield', 'Stratovolcano', 'Subglacial volcano', 'Submarine volcano', 'Volcanic field'] 
```

```python
for volcano_type in volcano_types:
    volcano_summary = wikipedia.WikipediaPage(title = volcano_type).summary
    #volcano_summary = volcano_summary.rstrip('\n').rstrip("\'").rstrip("\n\n")
    volcano_summary_list.append(volcano_summary)    
```

```python
volcano_summary_list
```

```python
volcano_summary_list = volcano_summary_list.replace('\n','').replace("\'","")
```

```python
volcano_dictionary = dict(zip(volcano_types, volcano_summary_list))
```

```python
volcano_dictionary.to_csv("Volcano_types.csv")
```

```python
import csv
csv_file = "volcano_data.csv"
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile)
        writer.writeheader()
        for data in volcano_dictionary:
            writer.writerow(data)
except:
    print("error")
```

```python
volcano_df = pd.DataFrame.from_dict(volcano_dictionary,orient='index')
```

```python
volcano_df.to_csv("volcano_data.csv")
```
